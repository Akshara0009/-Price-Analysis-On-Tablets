```python
brand=[]
color=[]
Ram=[]
Rom=[]
screensize=[]
battery=[]
Megapixel=[]
Rating=[]
Price=[]
        
for i in range(1,21):
    print(f"https://www.flipkart.com/search?q=Tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}")
    
    page=requests.get(url)
    
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("^[A-Za-z]+",text)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("\(([^()]+)\)$",text)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("(\d+)\sGB\sRAM",text)
        if regex:
            Ram.append(regex[0])
        else:
            Ram.append(np.nan)
```

```python
brand=[]
color=[]
Ram=[]
Rom=[]
screensize=[]
battery=[]
Megapixel=[]
Rating=[]
price=[]
        
for i in range(1,21):
    print(f"https://www.flipkart.com/search?q=Tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}")
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("^[A-Za-z]+",text)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("\(([^()]+)\)$",text)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("(\d+)\sGB\sRAM",text)
        if regex:
            Ram.append(regex[0])
        else:
            Ram.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("(\d+)\sGB\sROM",text)
        if regex:
            Rom.append(regex[0])
        else:
            Rom.append(np.nan)
        
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("(\d+(?:\.\d+)?)\s+[Ii]nch",text)
        if regex:
            screensize.append(regex[0])
        else:
            screensize.append(np.nan)
            
    for i in soup.find_all("div",class_="_2kHMtA"):
        text=i.text
        regex=re.findall("Battery:\s+(\d+)\s*\w+",text)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
            
    for i in soup.find_all("div",class_="_2kHMtA"):
        text=i.text
        regex=re.findall("(\d+)\s*MP",text)
        if regex:
            Megapixel.append(regex[0])
        else:
            Megapixel.append(np.nan)
            
    for i in soup.find_all("a",class_="_1fQZEK"):
        rat=i.find("div",class_="_3LWZlK")
        if rat:
            Rating.append(rat.text)
        else:
            Rating.append(np.nan)
            
    for i in soup.find_all("div",class_="_30jeq3 _1_WHN1"):
        price.append(i.text)
```
